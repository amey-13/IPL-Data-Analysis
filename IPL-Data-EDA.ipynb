### Total matches played, total wins, wins while chasing, wins while batting first for each team.

```python
# Total number of matches played by each team
total_matches_played = (matches_df['team1'].value_counts() + matches_df['team2'].value_counts()).rename_axis('Team').reset_index(name='Total_matches_played')

# Total wins by each team
total_wins = matches_df['winner'].value_counts().rename_axis('Team').reset_index(name='Total_wins')

# Wins while chasing
chasing_wins = matches_df[matches_df['toss_decision'] == 'field']['winner'].value_counts().rename_axis('Team').reset_index(name='Chasing_wins')

# Wins while batting
batting_first_wins = matches_df[matches_df['toss_decision'] == 'bat']['winner'].value_counts().rename_axis('Team').reset_index(name='Batting_wins')

# Merging all the dataframes
team_wins_merged = total_matches_played.merge(total_wins, on='Team', how='outer').merge(chasing_wins, on='Team', how='outer').merge(batting_first_wins, on='Team', how='outer').fillna(0)

# Sorting by total wins
team_wins_merged.sort_values(by='Total_wins', ascending=False, inplace=True)

# Resetting the index and sorting again
team_wins_merged.set_index('Team').sort_values('Total_wins', ascending=False)
```

### Pie chart to determine the contribution of each team in the total wins of the league throughout the seasons.

```python
plt.figure(figsize=(10,10))
plt.pie(team_wins_merged['Total_wins'],
        labels=team_wins_merged['Team'],
        autopct='%1.1f%%',
        startangle=0,
        explode=[0.05]*len(team_wins_merged),
        pctdistance=0.8,
        labeldistance=1.05
        )
plt.title('Contribution of each team in the total wins of the league')
plt.show()
```

### Getting the above data in terms of the percentage of total value.

```python
team_wins_percentage = team_wins_merged.copy()

# Percentage of total wins
team_wins_percentage['Total_wins'] = (team_wins_percentage['Total_wins'] / team_wins_percentage['Total_matches_played']) * 100

# Percentage of Chasing wins
team_wins_percentage['Chasing_wins'] = (team_wins_percentage['Chasing_wins'] / team_wins_merged['Total_wins']) * 100

# Percentage of Batting first wins
team_wins_percentage['Batting_wins'] = (team_wins_percentage['Batting_wins'] / team_wins_merged['Total_wins']) * 100

# Sorting by total wins
team_wins_percentage.set_index('Team').sort_values('Total_wins', ascending=False)
```

```python
# Sorting the data by total wins
sorted_team_wins_percentage = team_wins_percentage.sort_values(by='Total_wins', ascending=False)

# Melting the dataframe
melted_team_wins_percentage = sorted_team_wins_percentage.melt(
    id_vars=['Team'],
    value_vars=['Total_wins', 'Chasing_wins', 'Batting_wins'],
    var_name='Win_Type',
    value_name='Percentage'
)

# Plotting with the help of bar plot
plt.figure(figsize=(12,7))
sns.barplot(data=melted_team_wins_percentage, x='Team', y='Percentage', hue='Win_Type', palette='muted')

plt.title("Percentage wins by team")
plt.xlabel("Team")
plt.ylabel("Percentage")
plt.legend(title='Win Type', bbox_to_anchor=(1, 1), loc='upper left')
plt.xticks(rotation=90)
plt.tight_layout()
```

### Total Matches played, total toss wins, total matches won after winning the toss, total matches won after winning the toss and batting first and total matches won after winning the toss and fielding first

```python
# Total matches won
total_matches_won = (matches_df['team1'].value_counts() + matches_df['team2'].value_counts()).rename_axis('Team').reset_index(name='Total_matches_won')

# Total toss wins
toss_winners = matches_df['toss_winner'].value_counts().rename_axis('Team').reset_index(name='Total_toss_wins')

# Pivot table
team_wins_pivot = pd.pivot_table(
    data=matches_df,
    index='winner',
    columns='toss_decision',
    aggfunc='size',
    fill_value=0
).rename_axis('Team').reset_index().rename(columns={'bat':'Batting_first_win_after_toss_win', 'field':'Field_first_win_after_toss_win'})

# Matches of teams winning the toss and winning the match
toss_winner_matches = matches_df[matches_df['toss_winner'] == matches_df['winner']]['winner'].value_counts().rename_axis('Team').reset_index(name='Toss_won_match_won')

# Merging the 3 tables
team_toss_wins_merged = total_matches_won.merge(toss_winners, on='Team', how='outer').merge(toss_winner_matches, on='Team', how='outer').merge(team_wins_pivot, on='Team', how='outer').fillna(0)

# Sorting values
team_toss_wins_merged.sort_values(by='Toss_won_match_won', ascending=False, inplace=True)

# Resetting the index
team_toss_wins_merged.set_index('Team').sort_values('Toss_won_match_won', ascending=False)
```

### Total runs scored by each batting team in a particular match, melting the dataframe and resetting the index

### Teams scoring most runs in death (16-20) overs

```python
# Death overs
death_overs = deliveries_df[(deliveries_df['over'] >= 16) & (deliveries_df['over'] <= 20)]

team_runs = death_overs.groupby('batting_team')['total_runs'].sum().reset_index(name='death_overs_runs')

# Filtering the valid balls (excluding the runs scored in extras)
valid_balls = death_overs[~death_overs['extras_type'].isin(['wides', 'noballs'])]

# Count of balls faced by each team
team_balls_faced = valid_balls.groupby('batting_team')['ball'].count().reset_index(name='balls_faced')

# Merging balls faced and runs made
team_stats = pd.merge(team_runs, team_balls_faced, on='batting_team')

# Strike rate
team_stats['strike_rate'] = (team_stats['death_overs_runs'] / team_stats['balls_faced']) * 100

# Wickets lost
wickets_lost = death_overs[death_overs['is_wicket'] == 1].groupby('batting_team')['is_wicket'].count().reset_index(name='wickets_lost')

# Merging wickets lost and filling missing values
team_stats = pd.merge(team_stats, wickets_lost, on='batting_team', how='left').fillna(0)

# Sorting by strike rate
team_stats =  team_stats.sort_values(by='strike_rate', ascending=False).reset_index(drop=True)
team_stats
```

### Average of batters who played atleast 25 matches

```python
# Getting the relevant coulmns
batter_data = deliveries_df[['match_id', 'batter', 'batsman_runs']]

# Matches played per batter
batter_matches = batter_data.groupby('batter')['match_id'].nunique().reset_index(name='matches_played')

# Total runs scored per batter
batter_runs = batter_data.groupby('batter')['batsman_runs'].sum().reset_index(name='total_runs')

# Merging the data
batter_stats = pd.merge(batter_matches, batter_runs, on='batter')

# Average runs per match for each batter
batter_stats['avg_runs'] = batter_stats['total_runs'] / batter_stats['matches_played']

# Minimum 25 matches
batter_stats = batter_stats[batter_stats['matches_played'] >= 25]

# Sorting the data
batter_avg = batter_stats.sort_values(by='avg_runs', ascending=False).reset_index(drop=True)

batter_avg
```

### Most economical bowlers (minimum 80 overs bowled)

```python
# Runs conceded
runs_conceded = deliveries_df.groupby('bowler')['total_runs'].sum().reset_index(name='runs_conceded')

# Filtering extras
valid_balls = deliveries_df[~deliveries_df['extras_type'].isin(['wides', 'noballs'])]

# Count valid deliveries
balls_bowled = valid_balls.groupby('bowler')['ball'].count().reset_index(name='balls_bowled')
balls_bowled['overs_bowled'] = balls_bowled['balls_bowled'] / 6

# Merging the data
bowler_stats = pd.merge(runs_conceded, balls_bowled[['bowler', 'overs_bowled']], on='bowler', how='left')

# Economy rate
bowler_stats['economy_rate'] = bowler_stats['runs_conceded'] / bowler_stats['overs_bowled']

# Minimum 80 overs bowled
bowler_stats = bowler_stats[bowler_stats['overs_bowled'] >= 80]

# Sorting the data
most_economical_bowlers = bowler_stats.sort_values(by='economy_rate', ascending=True).reset_index(drop=True)

# Top 20 most economical bowlers
top_20_economical_bowlers = most_economical_bowlers.head(20)

top_20_economical_bowlers
```

```python
# Bar plot
plt.figure(figsize=(12,7))

sns.barplot(data=top_20_economical_bowlers, x='economy_rate', y='bowler', palette='magma')

for i, row in top_20_economical_bowlers.iterrows():
  plt.text(row.economy_rate + 0.02, i, round(row.economy_rate, 2), ha="left", va='center')

plt.title("Top 20 most economical bowlers (minimum 80 overs bowled)")
plt.xlabel('Economy Rate')
plt.ylabel('Bowler')
plt.tight_layout()
```
